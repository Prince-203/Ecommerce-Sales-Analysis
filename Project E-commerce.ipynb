# Visualization Of Data

# Bar Plot Between Product Name Vs Category

```python
plt.figure(figsize=(16,8))
plt.bar("Product Name", "Category", data=df, align="center", width=0.6, color='#e74c3c',edgecolor='k', linewidth=2)
plt.xticks(rotation=60)
plt.show()
```

```python
# Less Frequency Items: Blender, Microwave, Toaster, Vacuum Cleaner, Coffee Maker
# Medium Frequency Items: Laptop, Sneakers, Tablet, Hat, Headphones, Smartwatch, Smartphone, Jacket, T-shirt, Jeans
# High Frequency Items: Perfume, Fiction, Biography, Science Fiction, Shampoo, Non-Fiction, Lipstick, Mascara, Face Creame, Mystery
```

# Pie Plot Of Category

```python
plt.figure(figsize=(4,4))
e=[0.1,0,0.5,0,0]
c=['c','gold','r','y','grey']
df["Category"].value_counts().plot.pie(autopct="%0.1f%%", 
                                       shadow=True,
                                       radius = 1,
                                       startangle=180,
                                       explode=e,
                                       colors=c)
plt.axis('equal')
plt.show()
```

# Bar Plot Of Sales VS Profit

```python
plt.rcParams["figure.figsize"] = [10,8]
df.groupby('Category')[['Profit','Sales']].agg(["sum"]).plot.bar(color=['#3498db', '#1abc9c'], edgecolor='k', linewidth=2)
plt.title("Total Profit and Sales Per Category")
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# 1. Key Observations
# Sales values are much higher than profit values across all categories.
# The proportion of profit to sales appears relatively consistent.
```

# Count Plot Of Cities 

```python
print(df["State"].value_counts())
plt.figure(figsize=(20,8))
sns.countplot(x=df["State"])
plt.xticks(rotation=90,fontsize=15)
plt.show()
```

# Sales and Profit Region Wise

```python
plt.rcParams['figure.figsize']=[17,10]
ax = df.groupby('Region')[["Profit","Sales"]].agg(["sum"]).plot.bar()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=13)
    
plt.title("Total Profit and Sales Per Region")
plt.show()
```
